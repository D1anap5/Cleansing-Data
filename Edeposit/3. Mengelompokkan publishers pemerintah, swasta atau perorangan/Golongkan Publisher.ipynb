```python
#mengelompokkan nama publisher ke golongan swasta, pemerintah atau perorangan#
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile

df = pd.read_excel('publishers.xlsx')
df.loc[(df['name'].str.startswith('CV') | df['name'].str.startswith('PT') | df['name'].str.startswith('Yayasan'))  
       , 'Golongan'] = '1'
df.loc[(df['name'].str.startswith('BPS') | df['name'].str.startswith('DEPOSIT') | df['name'].str.startswith('Badan') |
       df['name'].str.startswith('Balai') | df['name'].str.startswith('Nasional') | df['name'].str.endswith('Nasional') | df['name'].str.startswith('Balitbang')
       | df['name'].str.startswith('Bank Indonesia') | df['name'].str.startswith('Bawaslu') |
        df['name'].str.startswith('Deputi') | df['name'].str.startswith('Dinas') | df['name'].str.startswith('Direktorat') 
        | df['name'].str.startswith('Kementerian')), 'Golongan'] = '3'
df['Golongan'].fillna('2', inplace = True)
df
#df.to_csv(r'C:\Users\diana\Downloads\publisher_golongan.csv',index=False)
```
